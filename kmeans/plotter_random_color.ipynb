```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=3, 
                  cluster_std=0.5, 
                  shuffle=True, 
                  random_state=0)
```

```python
print(X)
```

```python
print(y)
```

```python
print(X[:,0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], 
            c='white', marker='o', edgecolor='black', s=50)
plt.grid()
plt.tight_layout()
#plt.savefig('images/11_01.png', dpi=300)
plt.show()
```

```python
import random

counter = 0
for i in y:
    
    clstr = random.randint(0, 3)
    
    if(clstr == 0):
        plt.scatter(X[counter, 0], X[counter, 1], 
            c='orange', marker='o', edgecolor='orange', s=50)

    if(clstr == 1):
        plt.scatter(X[counter, 0], X[counter, 1], 
            c='lightblue', marker='o', edgecolor='lightblue', s=50)
    
    if(clstr == 2):
        plt.scatter(X[counter, 0], X[counter, 1], 
            c='lightgreen', marker='o', edgecolor='orange', s=50)
    
    counter = counter + 1
    
plt.grid()
plt.tight_layout()
#plt.savefig('images/11_01.png', dpi=300)
plt.show()    
```
